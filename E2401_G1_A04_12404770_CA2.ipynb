5. Create a story around year wise changes in the overall IPL
performances. For that, get the records of year wise total run scored,
total catches and total wickets taken, including all the player records.
Comment on the year-wise variations in them and support your
comments using proper visualizations.

```python
#seperating data file for year wise analysis
datafile_1 = pd.read_csv('C:/Users/Girip/OneDrive/Desktop/IPL2024_Data_Edited.csv')
df_5=datafile_1[['Year','Player_Name', 'Runs_Scored', 'Catches_Taken', 'Wickets_Taken']]
df_5
```

```python
#saving Data File
df_5.to_csv('Ques_5.csv')
```

            According to my analysis the runs in overall ipl was gradually increasing and the players are also well performing by giving the performance analysis as runs scored, wickets taken and catches taken. Rest of performance analysis were given below.

```python
#total runs in ipl year wise
import matplotlib.pyplot as plt
yearly_runs = df_5.groupby('Year')['Runs_Scored'].sum()
plt.figure(figsize=(12, 6))
yearly_runs.plot(kind='bar', color='b')
plt.title('Year-wise Total Runs in IPL')
plt.xlabel('Year')
plt.ylabel('Total Runs')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Year wise Total wickets taken in ipl 
yearly_runs = df_5.groupby('Year')['Catches_Taken'].sum()
plt.figure(figsize=(12, 6))
yearly_runs.plot(kind='bar', color='m')
plt.title('Year-wise Total Runs in IPL')
plt.xlabel('Year')
plt.ylabel('Total Wickets Taken')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Year wise Catches takenin ipl
yearly_runs = df_5.groupby('Year')['Catches_Taken'].sum()
plt.figure(figsize=(12, 6))
yearly_runs.plot(kind='bar', color='g')
plt.title('Year-wise Total Catches Taken in IPL')
plt.xlabel('Year')
plt.ylabel('Total Wickets Taken')
plt.grid(True)
plt.tight_layout()
plt.show()
```
